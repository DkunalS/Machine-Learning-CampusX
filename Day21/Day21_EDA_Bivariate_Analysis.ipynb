```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
iris = sns.load_dataset('iris')
iris
```

1. ScatterPlot (Numerical - Numerical)
    - When Data is Numerical




```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue="sex", style="smoker", size="size")
```

2. Bar Plot (Numerical- Categorical)
    - when data is numerical and categorical

```python
sns.barplot(x='pclass', y='age', data=titanic, hue='sex')
```

3. Box Plot (Numerical - Categorical)
  - When Data is Numerical and Categorical

```python
sns.boxplot(x='sex', y='age', data=titanic, hue='survived')
```

4. DistPlot(Numerical- Categorical)
  - When Data is Numerical- Categorical
  

```python
sns.displot(titanic[titanic['survived']== 0]['age'])
sns.displot(titanic[titanic['survived']== 1]['age'])
```

5. HeatMap
  - When data is categorical and categorical

```python
sns.heatmap(pd.crosstab(titanic['sex'], titanic['survived']))
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```
