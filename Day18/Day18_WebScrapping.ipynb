```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpages = requests.get('https://www.meesho.com/ethnicwear-women/pl/3tq', headers=headers).text
```

```python
soup = BeautifulSoup(webpages, 'lxml')
```

```python
soup.prettify()
df = pd.DataFrame()
```

```python
# Fetching name through Scrapping
name = [i.text for i in (soup.find_all('p', class_='sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU'))]
df['name'] = name
df
```

```python
# Fetching price through Scrapping
price = [i.text for i in (soup.find_all('h5'))]
df['price'] = price
df['price'] = df['price'].str.replace('₹', '', regex=False)
df
```

```python
# Fetching rating through Scrapping
rating = [i.text for i in (soup.find_all('span', class_='sc-eDvSVe laVOtN'))]
df['rating'] = rating
df
```

```python
# Fetching review counting through scrapping
reviews = [i.text for i in (soup.find_all('span', class_='sc-eDvSVe XndEO NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc'))]
df['reviews'] = reviews
df
```
