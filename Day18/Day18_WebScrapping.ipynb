```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpages = requests.get('https://www.meesho.com/ethnicwear-women/pl/3tq', headers=headers).text
```

```python
soup = BeautifulSoup(webpages, 'lxml')
```

```python
soup.prettify()
```

```python
length = len(soup.find_all('h5'))
for i in length:
    print(soup.find_all('h5')[i].text)
```
